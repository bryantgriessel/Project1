```python
#Imported Dependencies
from pathlib import Path
import pandas as pd
import scipy.stats as stats
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#File path for DataFrames
obesity_csv = pd.read_csv("../resources/Several_States_Complete.csv")
obesity_df = pd.DataFrame(obesity_csv)

#Bins for obesity score
bins = [0, 20, 40, 60, 80, 100]

# Categorize the data into the bins and create a column in the DataFrame
obesity_df['WalkScoreRange'] = pd.cut(obesity_df["WalkScore"], bins=bins, labels=["0-20", "21-40", "41-60", "61-80", "81-100"])

obesity_df.head()
```

```python
#Independent T-Test on Rate of Obesity by population
stats.ttest_ind(obesity_df.loc[obesity_df['State'] == "TX"].ObesityScore, obesity_df.loc[obesity_df['State'] == "NY"].ObesityScore, equal_var=False)
```

```python
#Independent T-Test on Rate of Obesity by population
stats.ttest_ind(obesity_df.loc[obesity_df['State'] == "FL"].ObesityScore, obesity_df.loc[obesity_df['State'] == "IL"].ObesityScore, equal_var=False)
```

```python
#Independent T-Test on Rate of Obesity by population
stats.ttest_ind(obesity_df.loc[obesity_df['State'] == "AZ"].ObesityScore, obesity_df.loc[obesity_df['State'] == "OH"].ObesityScore, equal_var=False)
```

```python
# Scatter Plot of Data
x = len(obesity_df.loc[obesity_df['State'] == "TX"].ObesityScore)
x2 = len(obesity_df.loc[obesity_df['State'] == "NY"].ObesityScore)
plt.subplot(2, 1, 1)
plt.scatter(range(x), obesity_df.loc[obesity_df['State'] == "TX"].ObesityScore, label="Texas", alpha = .5)
plt.scatter(range(x2), obesity_df.loc[obesity_df['State'] == "NY"].ObesityScore,label ="New York", alpha = .5)
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(obesity_df.loc[obesity_df['State'] == "TX"].ObesityScore, 20, alpha=0.7, label="Texas")
plt.hist(obesity_df.loc[obesity_df['State'] == "NY"].ObesityScore, 20, alpha=0.7, label="New York")
plt.axvline(obesity_df.loc[obesity_df['State'] == "TX"].ObesityScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(obesity_df.loc[obesity_df['State'] == "NY"].ObesityScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend() 
plt.show()
```

```python
#Settings groups to perform an ANOVA test.
group0 = state1[state1["WalkScoreRange"] == "0-20"]["ObesityScore"]
group1 = state1[state1["WalkScoreRange"] == "21-40"]["ObesityScore"]
group2 = state1[state1["WalkScoreRange"] == "41-60"]["ObesityScore"]
group3 = state1[state1["WalkScoreRange"] == "61-80"]["ObesityScore"]
group4 = state1[state1["WalkScoreRange"] == "81-100"]["ObesityScore"]
# Perform the ANOVA test
stats.f_oneway(group0, group1, group2, group3, group4)
```
