```python
#Imported Dependencies
from pathlib import Path
import pandas as pd
import scipy.stats as stats
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#File path for DataFrames

obesity_csv = pd.read_csv("../resources/obesity_data.csv")

obesity_df = pd.DataFrame(obesity_csv)
obesity_df.head()
```

```python
#Setting two populations to perform an independent T-Test
population1 = obesity_df.loc[obesity_df['StateAbbr'] == "TX"].dropna(how='any')
population2 = obesity_df.loc[obesity_df['StateAbbr'] == "NY"].dropna(how='any')
Lat
nydata=population2[["StateAbbr","CityName","UniqueID","TractFIPS","Data_Value","Lat","Lon"]]
```

```python
#Gathering the count to make sure populations are similar size
len(population1.Data_Value)
#population1.count()
```

```python
#Gathering the count to make sure populations are similar size
#population2.Data_Value.count()
population2.count()
```

```python
# Scatter Plot of Data
x = len(population1.Data_Value)
x2 = len(population2.Data_Value)
plt.subplot(2, 1, 1)
plt.scatter(range(x), population1.Data_Value, label="Texas", alpha = .5)
plt.scatter(range(x2), population2.Data_Value, label="New York", alpha = .5)
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(population1.Data_Value, 20, alpha=0.7, label="Texas")
plt.hist(population2.Data_Value, 20, alpha=0.7, label="New York")
plt.axvline(population1.Data_Value.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(population2.Data_Value.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend() 
plt.show()
```

```python
#Calculating the means of the populations
population1.Data_Value.mean()
```

```python
#Calculating the means of the populations
population2.Data_Value.mean()
```

```python
#Independent T-Test on Rate of Obesity by population
stats.ttest_ind(population1.Data_Value, population2.Data_Value, equal_var=False)
```
