```python
#Imported Dependencies
from pathlib import Path
import pandas as pd
import scipy.stats as stats
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#File path for DataFrames
obesity_csv = pd.read_csv("../resources/Several_States_Complete.csv")
obesity_df = pd.DataFrame(obesity_csv)

#Bins for obesity score
bins = [0, 20, 40, 60, 80, 100]

# Categorize the data into the bins and create a column in the DataFrame
obesity_df['WalkScoreRange'] = pd.cut(obesity_df["WalkScore"], bins=bins, labels=["0-20", "21-40", "41-60", "61-80", "81-100"])

obesity_df.head()
```

```python
obesity_df["State"].value_counts()
```

```python
#Setting two populations to perform an independent T-Test
population1 = obesity_df.loc[obesity_df['State'] == "TX"].dropna(how='any')
population2 = obesity_df.loc[obesity_df['State'] == "NY"].dropna(how='any')
```

```python
#Gathering the count to make sure populations are similar size
len(population1.ObesityScore)
#population1.count()
```

```python
#Gathering the count to make sure populations are similar size
#population2.Data_Value.count()
population2.count()
```

```python
# Scatter Plot of Data
x = len(population1.ObesityScore)
x2 = len(population2.ObesityScore)
plt.subplot(2, 1, 1)
plt.scatter(range(x), population1.ObesityScore, label="Texas", alpha = .5)
plt.scatter(range(x2), population2.ObesityScore, label="New York", alpha = .5)
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(population1.ObesityScore, 20, alpha=0.7, label="Texas")
plt.hist(population2.ObesityScore, 20, alpha=0.7, label="New York")
plt.axvline(population1.ObesityScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(population2.ObesityScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend() 
plt.show()
```

```python
#Calculating the means of the populations
population1.ObesityScore.mean()
```

```python
#Calculating the means of the populations
population2.ObesityScore.mean()
```

```python
#Independent T-Test on Rate of Obesity by population
stats.ttest_ind(population1.ObesityScore, population2.ObesityScore, equal_var=False)
```

```python
#Setting two populations to perform an independent T-Test
population1 = obesity_df.loc[obesity_df['State'] == "TX"].dropna(how='any')
population2 = obesity_df.loc[obesity_df['State'] == "NY"].dropna(how='any')
```

```python
# Scatter Plot of Walkability Data
x = len(population1.WalkScore)
x2 = len(population2.WalkScore)
plt.subplot(2, 1, 1)
plt.scatter(range(x), population1.WalkScore, label="Texas", alpha = .5)
plt.scatter(range(x2), population2.WalkScore, label="New York", alpha = .5)
plt.legend()

# Histogram Plot of Data
plt.subplot(2, 1, 2)
plt.hist(population1.WalkScore, 20, alpha=0.7, label="Texas")
plt.hist(population2.WalkScore, 20, alpha=0.7, label="New York")
plt.axvline(population1.WalkScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(population2.WalkScore.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend()
plt.show()
```

```python
#Box plot that will show obesity score means in each walk score range
population1.boxplot("ObesityScore", by="WalkScoreRange", figsize=(20, 10),grid=False)
plt.show()
```

```python
#Settings groups to perform an ANOVA test.
group0 = population1[population1["WalkScoreRange"] == "0-20"]["ObesityScore"]
group1 = population1[population1["WalkScoreRange"] == "21-40"]["ObesityScore"]
group2 = population1[population1["WalkScoreRange"] == "41-60"]["ObesityScore"]
group3 = population1[population1["WalkScoreRange"] == "61-80"]["ObesityScore"]
group4 = population1[population1["WalkScoreRange"] == "81-100"]["ObesityScore"]
```

```python
# Perform the ANOVA test
stats.f_oneway(group0, group1, group2, group3, group4)
```
